# Main Evaluation: Compute Average Metrics (No Full Audio Storage)

## Currently only 16 samples used as System RAM was getting exhausted otherwise

```python
import os
import random
import shutil
import time
import torch
import torchaudio
import numpy as np
from audioseal import AudioSeal
from pesq import pesq
from pystoi import stoi
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from scipy.signal import resample

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

dataset_link = dataset_link

# Sample Extraction

def list_audio_files(root_dir, file_extension='.wav'):
    audio_files = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.lower().endswith(file_extension):
                audio_files.append(os.path.join(root, file))
    return audio_files

def extract_samples(source_dir, dest_dir, n_samples=5, file_extension='.wav'):
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
    audio_files = list_audio_files(source_dir, file_extension)
    if len(audio_files) == 0:
        print(f"No files with extension {file_extension} found in {source_dir}")
        return []
    selected_files = random.sample(audio_files, min(n_samples, len(audio_files)))
    for file_path in selected_files:
        file_name = os.path.basename(file_path)
        dest_path = os.path.join(dest_dir, file_name)
        shutil.copy(file_path, dest_path)
    print(f"Extracted {len(selected_files)} sample files from:\n  {source_dir}\nto:\n  {dest_path}")
    return selected_files

#  AudioSeal Inference and Metrics

def load_and_preprocess(audio_path, target_sr=16000):
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"Audio file not found: {audio_path}")
    wav, sr = torchaudio.load(audio_path)
    if sr != target_sr:
        resampler = torchaudio.transforms.Resample(sr, target_sr)
        wav = resampler(wav)
        sr = target_sr
    return wav, sr

def compute_si_snr(clean, degraded):
    clean = clean.flatten()
    degraded = degraded.flatten()
    alpha = torch.dot(degraded, clean) / (torch.norm(clean)**2 + 1e-10)
    projection = alpha * clean
    noise = degraded - projection
    si_snr = 10 * torch.log10((torch.norm(projection)**2) / (torch.norm(noise)**2 + 1e-10))
    return si_snr.item()

def compute_pesq_score(clean, degraded, sr=16000):
    return pesq(sr, clean.detach().numpy(), degraded.detach().numpy(), 'wb')

def compute_stoi_score(clean, degraded, sr=16000):
    return stoi(clean.detach().numpy(), degraded.detach().numpy(), sr, extended=False)

def process_audio_with_masking(audio_path, generator, detector, gt_message=None):
    wav, sr = load_and_preprocess(audio_path, target_sr=16000)
    if wav.dim() == 2:
        wav = wav.unsqueeze(0)

    t0 = time.time()
    if gt_message is not None:
        watermark = generator.get_watermark(wav, sr, message=gt_message)
    else:
        watermark = generator.get_watermark(wav, sr)
    gen_time = (time.time() - t0) * 1000  # ms
    watermarked = wav + watermark

    t0 = time.time()
    det_score, decoded_message = detector.detect_watermark(watermarked, sr)
    det_time = (time.time() - t0) * 1000  # ms

    orig_mono = wav.squeeze(0).mean(dim=0)
    watermarked_mono = watermarked.squeeze(0).mean(dim=0)
    si_snr_val = compute_si_snr(orig_mono, watermarked_mono)
    try:
        pesq_val = compute_pesq_score(orig_mono, watermarked_mono, sr)
    except Exception as e:
        pesq_val = np.nan
        print(f"PESQ computation failed for {audio_path}: {e}")
    try:
        stoi_val = compute_stoi_score(orig_mono, watermarked_mono, sr)
    except Exception as e:
        stoi_val = np.nan
        print(f"STOI computation failed for {audio_path}: {e}")

    if gt_message is not None and decoded_message is not None:
        gt_msg_np = gt_message.cpu().numpy().flatten()
        dec_msg_np = (decoded_message.cpu().numpy().flatten() > 0.5).astype(int)
        ham_dist = np.mean(gt_msg_np != dec_msg_np)
        attr_correct = ham_dist < 0.1
    else:
        ham_dist = None
        attr_correct = False

    total_samples = watermarked.shape[-1]
    total_time = total_samples / sr
    mask_fraction = random.uniform(0.1, 0.5)
    mask_duration = mask_fraction * total_time
    start_time = random.uniform(0, total_time - mask_duration)
    end_time = start_time + mask_duration
    start_idx = int(start_time * sr)
    end_idx = int(end_time * sr)
    masked_audio = watermarked.clone()
    masked_audio[:, :, start_idx:end_idx] = wav[:, :, start_idx:end_idx]

    masked_mono = masked_audio.squeeze(0).mean(dim=0)
    masked_si_snr = compute_si_snr(orig_mono, masked_mono)
    try:
        masked_pesq = compute_pesq_score(orig_mono, masked_mono, sr)
    except Exception as e:
        masked_pesq = np.nan
        print(f"PESQ (masked) computation failed for {audio_path}: {e}")
    try:
        masked_stoi = compute_stoi_score(orig_mono, masked_mono, sr)
    except Exception as e:
        masked_stoi = np.nan
        print(f"STOI (masked) computation failed for {audio_path}: {e}")
    masked_det_score, _ = detector.detect_watermark(masked_audio, sr)

    # Return scalar metrics only (to save memory).
    return {
        "si_snr": si_snr_val,
        "pesq": pesq_val,
        "stoi": stoi_val,
        "gen_time_ms": gen_time,
        "det_time_ms": det_time,
        "detector_score": det_score,
        "mask_start_sec": start_time,
        "mask_end_sec": end_time,
        "masked_si_snr": masked_si_snr,
        "masked_pesq": masked_pesq,
        "masked_stoi": masked_stoi,
        "masked_detector_score": masked_det_score,
        "hamming_distance": ham_dist,
        "attribution_correct": attr_correct
    }

#  Main Evaluation (Average Metrics Computation)

def main_evaluation():
    sample_files = extract_samples(dataset_link, "/content/sample_extracted_audio", n_samples=16)
    if len(sample_files) == 0:
        print("No sample files extracted. Check your dataset path.")
        return

    print("\nLoading AudioSeal models...")
    generator = AudioSeal.load_generator("audioseal_wm_16bits")
    detector = AudioSeal.load_detector("audioseal_detector_16bits")

    gt_message = torch.randint(0, 2, (1, generator.msg_processor.nbits), dtype=torch.int)

    global eval_metrics, sample_files_global
    eval_metrics = []  # Scalar metrics only
    sample_files_global = sample_files  # Save file paths for visualization

    print("\nProcessing samples for average metrics...")
    for sample in sample_files:
        metrics = process_audio_with_masking(sample, generator, detector, gt_message=gt_message)
        eval_metrics.append(metrics)
        print(f"Processed: {sample}")
        print(f"  SI-SNR (Watermarked): {metrics['si_snr']:.2f} dB, Detector Score: {metrics['detector_score']:.3f}")
        print(f"  SI-SNR (Masked): {metrics['masked_si_snr']:.2f} dB, Detector Score (Masked): {metrics['masked_detector_score']:.3f}")
        print(f"  Masking Interval: {metrics['mask_start_sec']:.2f} to {metrics['mask_end_sec']:.2f} seconds")

    si_snr_vals = np.array([m['si_snr'] for m in eval_metrics])
    pesq_vals = np.array([m['pesq'] for m in eval_metrics])
    stoi_vals = np.array([m['stoi'] for m in eval_metrics])
    det_scores = np.array([m['detector_score'] for m in eval_metrics])
    masked_si_snr_vals = np.array([m['masked_si_snr'] for m in eval_metrics])
    masked_pesq_vals = np.array([m['masked_pesq'] for m in eval_metrics])
    masked_stoi_vals = np.array([m['masked_stoi'] for m in eval_metrics])
    masked_det_scores = np.array([m['masked_detector_score'] for m in eval_metrics])
    ham_dists = np.array([m['hamming_distance'] for m in eval_metrics if m['hamming_distance'] is not None])

    avg_si_snr = np.nanmean(si_snr_vals)
    avg_pesq = np.nanmean(pesq_vals)
    avg_stoi = np.nanmean(stoi_vals)
    avg_det_score = np.mean(det_scores)
    avg_masked_si_snr = np.nanmean(masked_si_snr_vals)
    avg_masked_pesq = np.nanmean(masked_pesq_vals)
    avg_masked_stoi = np.nanmean(masked_stoi_vals)
    avg_masked_det_score = np.mean(masked_det_scores)
    avg_hamming = np.nanmean(ham_dists) if len(ham_dists) > 0 else None

    print("\n===== Average Metrics Across Samples =====")
    print("Watermarked Audio Metrics:")
    print(f"  Average SI-SNR: {avg_si_snr:.2f} dB")
    print(f"  Average PESQ: {avg_pesq:.2f}")
    print(f"  Average STOI: {avg_stoi:.3f}")
    print(f"  Average Detector Score: {avg_det_score:.3f}")
    print("Masked Audio Metrics:")
    print(f"  Average SI-SNR: {avg_masked_si_snr:.2f} dB")
    print(f"  Average PESQ: {avg_masked_pesq:.2f}")
    print(f"  Average STOI: {avg_masked_stoi:.3f}")
    print(f"  Average Detector Score: {avg_masked_det_score:.3f}")
    if avg_hamming is not None:
        print(f"  Average Attribution Hamming Distance: {avg_hamming:.3f}")
    else:
        print("  Attribution Hamming Distance: N/A")

if __name__ == "__main__":
    main_evaluation()
```

# Visualization and Playback

```python
import matplotlib.pyplot as plt
from IPython.display import Audio, display
import numpy as np
import random
from audioseal import AudioSeal

# Check if sample_files_global exists.
if 'sample_files_global' not in globals():
    raise NameError("The variable 'sample_files_global' is not defined. Please run the main evaluation cell first.")

# Let the user choose a sample index (0-based) for visualization.
sample_index = int(input("Enter the sample index (0 to {}): ".format(len(sample_files_global) - 1)))
sample_path = sample_files_global[sample_index]
print(f"Visualizing sample: {sample_path}")

# Define a helper function to re-process a single sample with full audio data.
def process_full_audio(audio_path, generator, detector, gt_message=None):
    wav, sr = load_and_preprocess(audio_path, target_sr=16000)
    if wav.dim() == 2:
        wav = wav.unsqueeze(0)
    watermark = generator.get_watermark(wav, sr, message=gt_message) if gt_message is not None else generator.get_watermark(wav, sr)
    watermarked = wav + watermark
    total_samples = watermarked.shape[-1]
    total_time = total_samples / sr
    mask_fraction = random.uniform(0.1, 0.5)
    mask_duration = mask_fraction * total_time
    start_time = random.uniform(0, total_time - mask_duration)
    end_time = start_time + mask_duration
    start_idx = int(start_time * sr)
    end_idx = int(end_time * sr)
    masked_audio = watermarked.clone()
    masked_audio[:, :, start_idx:end_idx] = wav[:, :, start_idx:end_idx]
    return {"original_audio": wav, "watermarked_audio": watermarked, "masked_audio": masked_audio, "sr": sr}

# Load models if not defined.
if 'generator' not in globals():
    generator = AudioSeal.load_generator("audioseal_wm_16bits")
if 'detector' not in globals():
    detector = AudioSeal.load_detector("audioseal_detector_16bits")
gt_message = torch.randint(0, 2, (1, generator.msg_processor.nbits), dtype=torch.int)

# Re-process the selected sample for full audio data.
vis_metrics = process_full_audio(sample_path, generator, detector, gt_message=gt_message)
sr = vis_metrics["sr"]
original_audio = vis_metrics["original_audio"]
watermarked_audio = vis_metrics["watermarked_audio"]
masked_audio = vis_metrics["masked_audio"]
watermark_audio = watermarked_audio - original_audio

# Convert tensors to mono numpy arrays.
orig_np = original_audio.squeeze(0).mean(dim=0).detach().numpy()
watermarked_np = watermarked_audio.squeeze(0).mean(dim=0).detach().numpy()
masked_np = masked_audio.squeeze(0).mean(dim=0).detach().numpy()
watermark_np = watermark_audio.squeeze(0).mean(dim=0).detach().numpy()

# Compute common y-axis limits.
global_min = min(orig_np.min(), watermarked_np.min(), masked_np.min(), watermark_np.min())
global_max = max(orig_np.max(), watermarked_np.max(), masked_np.max(), watermark_np.max())

def plot_waveform(signal, sr, title, ymin, ymax):
    plt.figure(figsize=(10, 4))
    times = np.linspace(0, len(signal)/sr, num=len(signal))
    plt.plot(times, signal)
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.ylim([ymin, ymax])
    plt.tight_layout()
    plt.show()

print("Plotting waveforms with common y-axis scale...")
plot_waveform(orig_np, sr, "Original Audio Waveform", global_min, global_max)
plot_waveform(watermarked_np, sr, "Watermarked Audio Waveform", global_min, global_max)
plot_waveform(masked_np, sr, "Masked Watermarked Audio Waveform", global_min, global_max)
plot_waveform(watermark_np, sr, "Extracted Watermark Waveform (Watermarked - Original)", global_min, global_max)

# Get per-frame detector output for the masked audio.
det_output_masked, _ = detector(masked_audio, sr)
det_probs_masked = det_output_masked[0, 1, :].detach().cpu().numpy()

plt.figure(figsize=(10, 3))
times = np.linspace(0, len(orig_np)/sr, num=len(det_probs_masked))
plt.plot(times, det_probs_masked)
plt.xlabel("Time (s)")
plt.ylabel("Watermark Probability")
plt.title("Detector Output over Time (Masked Audio)")
plt.ylim([0, 1])
plt.tight_layout()
plt.show()

print("Playing Original Audio:")
display(Audio(orig_np, rate=sr))
print("Playing Watermarked Audio:")
display(Audio(watermarked_np, rate=sr))
print("Playing Masked Watermarked Audio:")
display(Audio(masked_np, rate=sr))
print("Playing Extracted Watermark Audio (Watermarked - Original):")
display(Audio(watermark_np, rate=sr))
```
